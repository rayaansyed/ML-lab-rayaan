```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sklearn modules
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    roc_curve, auc, RocCurveDisplay
)

# --- 1. Data Loading and Detailed Dataset Analysis ---

print("--- 1. Data Loading and Detailed Dataset Analysis ---")
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
feature_names = cancer.feature_names
target_names = cancer.target_names

# Create DataFrame for easy analysis
df = pd.DataFrame(X, columns=feature_names)
df['Diagnosis'] = y

# Display dataset dimensions and target distribution
print(f"Dataset Shape (Rows, Features): {df.shape}")

print("\n--- Target Distribution ---")
target_counts = df['Diagnosis'].value_counts()
target_table = pd.DataFrame({
    'Diagnosis': ['Benign (0)', 'Malignant (1)'],
    'Count': target_counts.values,
    'Percentage': (target_counts.values / len(df) * 100).round(2)
})
print(target_table.to_markdown(index=False))

print("\n--- Descriptive Statistics of Key Features ---")
key_features = ['mean radius', 'worst area', 'mean smoothness']
print(df[key_features].describe().to_markdown())

# Visualize Target Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Diagnosis', data=df, palette='viridis')
plt.title('Distribution of Diagnosis (0=Benign, 1=Malignant)')
plt.xticks([0, 1], ['Benign', 'Malignant'])
plt.show() #

# Split and Scale
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
print("\nData preparation complete.")

# --- 2. Model Training and Visualization ---

print("\n--- 2. Decision Tree and Random Forest Training ---")
# A. Decision Tree
dt_classifier = DecisionTreeClassifier(max_depth=5, random_state=42)
dt_classifier.fit(X_train_scaled, y_train)
dt_predictions = dt_classifier.predict(X_test_scaled)
dt_proba = dt_classifier.predict_proba(X_test_scaled)[:, 1]

# B. Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42, n_jobs=-1)
rf_classifier.fit(X_train_scaled, y_train)
rf_predictions = rf_classifier.predict(X_test_scaled)
rf_proba = rf_classifier.predict_proba(X_test_scaled)[:, 1]

# Visualize the Decision Tree
plt.figure(figsize=(25, 10))
plot_tree(
    dt_classifier, filled=True, feature_names=feature_names,
    class_names=target_names, rounded=True, fontsize=10
)
plt.title("Decision Tree Classifier Visualization (Max Depth = 5)")
plt.show() #

# --- 3. Enhanced Performance Evaluation ---

def evaluate_model_full(model_name, y_true, y_pred, y_proba):
    print(f"\n" + "="*70)
    print(f"*** {model_name} Performance Metrics ***")

    # Classification Report
    report = classification_report(y_true, y_pred, target_names=target_names, output_dict=True)
    report_df = pd.DataFrame(report).transpose().round(4)
    print("\nClassification Report Table:")
    print(report_df.to_markdown())

    # Confusion Matrix Visualization
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(
        cm, annot=True, fmt='d', cmap='Blues',
        xticklabels=target_names, yticklabels=target_names
    )
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show() #

    # ROC Curve
    fpr, tpr, thresholds = roc_curve(y_true, y_proba)
    roc_auc = auc(fpr, tpr)
    print(f"Area Under ROC Curve (AUC): {roc_auc:.4f}")
    return fpr, tpr, roc_auc

dt_fpr, dt_tpr, dt_auc = evaluate_model_full("Decision Tree", y_test, dt_predictions, dt_proba)
rf_fpr, rf_tpr, rf_auc = evaluate_model_full("Random Forest", y_test, rf_predictions, rf_proba)


# --- 4. Comparative ROC Curve Diagram ---

plt.figure(figsize=(8, 6))
plt.plot(dt_fpr, dt_tpr, color='red', lw=2, label=f'Decision Tree (AUC = {dt_auc:.4f})')
plt.plot(rf_fpr, rf_tpr, color='blue', lw=2, label=f'Random Forest (AUC = {rf_auc:.4f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity/Recall)')
plt.title('Receiver Operating Characteristic (ROC) Curve Comparison')
plt.legend(loc="lower right")
plt.grid(True)
plt.show() #

# --- 5. Feature Distribution Visualization ---

# Identify top features based on the Random Forest (a robust method)
feature_importances = pd.Series(
    rf_classifier.feature_importances_,
    index=feature_names
).sort_values(ascending=False)

top_4_features = feature_importances.index[:4]

plt.figure(figsize=(15, 10))
for i, feature in enumerate(top_4_features):
    plt.subplot(2, 2, i + 1)
    sns.histplot(
        data=df,
        x=feature,
        hue='Diagnosis',
        kde=True,
        palette={0: 'green', 1: 'red'},
        stat='density',
        common_norm=False
    )
    plt.title(f'Distribution of {feature} by Diagnosis')
    plt.legend(title='Diagnosis', labels=['Malignant', 'Benign'])

plt.tight_layout()
plt.show() #
```
