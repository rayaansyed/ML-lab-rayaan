```python
# ---------------------------------------------------------
# CLUSTERING ON HEALTHCARE DATASET
# K-Means • Spectral • Hierarchical Clustering
# ---------------------------------------------------------

# Install dependencies (optional, usually not needed in Colab)
# !pip install scikit-learn matplotlib pandas numpy

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, SpectralClustering, AgglomerativeClustering
import matplotlib.pyplot as plt

# -----------------------------
# 1. CREATE SYNTHETIC HEALTHCARE DATASET
# -----------------------------
np.random.seed(42)
n = 300  # number of samples

data = pd.DataFrame({
    "Age": np.random.normal(50, 12, n).clip(18, 90),
    "BMI": np.random.normal(27, 5, n).clip(15, 45),
    "Blood_Pressure": np.random.normal(120, 15, n).clip(80, 180),
    "Cholesterol": np.random.normal(200, 25, n).clip(120, 300)
})

# Standardizing data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# -----------------------------
# 2. APPLY CLUSTERING ALGORITHMS
# -----------------------------

# K-MEANS
kmeans = KMeans(n_clusters=3, random_state=42)
data["KMeans_Cluster"] = kmeans.fit_predict(scaled_data)

# SPECTRAL CLUSTERING
spectral = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=42)
data["Spectral_Cluster"] = spectral.fit_predict(scaled_data)

# HIERARCHICAL CLUSTERING
hierarchical = AgglomerativeClustering(n_clusters=3)
data["Hierarchical_Cluster"] = hierarchical.fit_predict(scaled_data)

# Preview dataset
print("Sample of Dataset with Clusters:")
print(data.head(), "\n")

# -----------------------------
# 3. VISUALIZATIONS
# -----------------------------
def plot_clusters(x, y, labels, title):
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, c=labels)
    plt.xlabel("Age")
    plt.ylabel("BMI")
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_clusters(data["Age"], data["BMI"], data["KMeans_Cluster"], "K-Means Clustering")
plot_clusters(data["Age"], data["BMI"], data["Spectral_Cluster"], "Spectral Clustering")
plot_clusters(data["Age"], data["BMI"], data["Hierarchical_Cluster"], "Hierarchical Clustering")

# -----------------------------
# 4. OPTIONAL: CLUSTER COUNT SUMMARY
# -----------------------------
print("\nCluster Distribution:")
print("K-Means:", data["KMeans_Cluster"].value_counts().to_dict())
print("Spectral:", data["Spectral_Cluster"].value_counts().to_dict())
print("Hierarchical:", data["Hierarchical_Cluster"].value_counts().to_dict())
```
